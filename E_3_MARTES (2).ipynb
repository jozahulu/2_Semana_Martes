```python
import numpy as np
import matplotlib.pyplot as plt
####-----------------###########
#PUEDE MODIFICAR LOS DATOS DE LA MANERA QUE QUIERAS
```

```python
N=22    #NUMERO DE DATOS ALEATORIOS
L=10    #LONGITUD DE DISTANCIA
datos=L*np.random.rand(N,2)
#x=np.linspace(0.,N,11)
plt.plot(datos,'.',color='red')
plt.title('Dispercion de Puntos')
```

```python
aleatorios = L*np.random.rand(N,2)
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
dtot = N*(N-1)/2
j = 0
dold = 0
dmax = 0
dis_datos = []
dis_aleatorios = []
z=[]
n=0
```

```python
for i in range(N):
    for j in range(i+1,N):
        if( j==i):
         n=0
        else:
         n=n+1
         d_d = distancia(datos[i][0],datos[j][0],datos[i][1],datos[j][1])
         d_a = distancia(aleatorios[i][0],aleatorios[j][0],aleatorios[i][1],aleatorios[j][1])
         dis_datos.append(d_d)
         dis_aleatorios.append(d_a)
         z.append(n)
         n=n
         #print(z)
```

```python
data_n, data_bin,m = plt.hist(dis_datos,bins=20,color='red')
print(len(dis_datos))
d=np.zeros(n)
print(len(d))
```

```python
ran_n, ran_bin, p = plt.hist(dis_aleatorios,bins=20)
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111, projection='3d')
xpos=dis_datos
ypos=dis_aleatorios
zpos=np.zeros(n)
dx=np.ones(n)
dy=np.ones(n)
dz=z
ax1.bar3d(xpos,ypos,zpos,dx,dy,dz,color='#00ceaa')
plt.title('GRAFICA 3D')
plt.ylabel('Distancia de DATOS')
plt.xlabel('Distancia de ALEATOTIA')
plt.show()
```
